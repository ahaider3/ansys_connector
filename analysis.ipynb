```python
import pyansys
```

```python
import numpy as np
```

```python
result = pyansys.ResultReader(input_path)
```

```python
stress = result.NodalStress(0)
```

```python
flat_stress = np.reshape(stress, (len(stress[0]) * len(stress) ,))
```

```python
flat_stress
```

```python
plt.plot(flat_stress)
```

```python
plt.ylabel("Nodal Stress")
plt.xlabel("Cell ID")
```

```python
plt.show()
```

```python
maxes = np.amax(stress, axis=1)
```

```python
plt.scatter(list(range(len(maxes))), maxes)
plt.xlabel("Row ID")
plt.ylabel("Max Stress by X dimension")
```

```python

plt.show()
```
